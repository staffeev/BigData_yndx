```python
import pandas as pd
```

```python
df1 = pd.read_csv("FIFA22_broken.csv")
#df1["ID"] = df1["ID"].astype(int)
df2 = pd.read_csv("FIFA23_broken.csv")
#df2["ID"] = df2["ID"].astype(int)
```

```python
def process(df1, df2):
    com = df1[df1["ID"].isin(df2["ID"])]
    res = com.set_index("ID").fillna(df2.set_index("ID"))
    res2 = res[~res.isna().any(axis=1)].reset_index().sort_values("ID")
    res2["ID"] = res2.pop("ID")
    return res2
```

```python
process(df1, df2)
```
