```python
import pandas as pd
```

```python
df = pd.read_csv("users_recency_monetary_frequency.csv")
```

```python
df
```

```python
def process(df):
    df["Recency_rank"] = df["Recency"].rank(ascending=False)
    df["Frequency_rank"] = df["Frequency"].rank(ascending=True)
    df["Monetary_rank"] = df["Monetary"].rank(ascending=True)
    df["Recency_rank"] /= df["Recency_rank"].max()
    df["Frequency_rank"] /= df["Frequency_rank"].max()
    df["Monetary_rank"] /= df["Monetary_rank"].max()
    df["RFM_Score"] = 5 * (
        15 * df["Recency_rank"] + 28 * df["Frequency_rank"] + 57 * df["Monetary_rank"]) / 100
    return df[(df["RFM_Score"] >= 4) & (df["RFM_Score"] <= 4.5)]["Customer_ID"].to_list()
```

```python
process(df)
```
