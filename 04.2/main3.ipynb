```python
import pandas as pd
```

```python
def process(df):
    df["Date"] = pd.to_datetime(df["Date"], format="%d/%m/%Y")
    f1 = df.groupby("Customer_ID").agg(
        Monetary=("Sales_Amount", "sum"), Recency=("Date", "max"), 
        Frequency=("Transaction_ID", "count")).reset_index()
    f1["Recency"] = (df["Date"].max() - f1["Recency"]).dt.days
    return f1[["Customer_ID", "Recency", "Frequency", "Monetary"]]
```

```python
df = pd.read_csv("customer_transactions_log.csv")
```

```python
process(df).to_csv("users_recency_monetary_frequency.csv")
```
