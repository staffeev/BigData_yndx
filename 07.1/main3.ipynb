```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("eplleaguetables.csv")
```

```python
plt.figure(figsize=(15, 10))
ax = plt.subplot(111)
df[df["Position"] == 1]["Points"].plot(kind="bar", stacked=True, color="#C9F76F", ax=ax, label="1-3 место")
df[df["Position"] == 3]["Points"].plot(kind="bar", stacked=True, color="#5CCCCC", ax=ax, label="_place")
df[df["Position"] == 19]["Points"].plot(kind="bar", stacked=True, color="#FF7373", ax=ax, label="19-20 место")
df[df["Position"] == 20]["Points"].plot(kind="bar", stacked=True, color="#FFFFFF", ax=ax, label="_place",
                                        alpha=0.7)
ax.set_xticklabels(df["Season"].drop_duplicates(), rotation=0)
ax.set_xlabel("Сезон, гг")
ax.set_ylabel("Очки")
ax.set_yticks(list(range(df["Points"].min(), df["Points"].max(), 3)) + [df["Points"].max()])
ax.set_axisbelow(True)
ax.legend()
ax.grid(axis="y", linestyle="--")
ax.set_title("Распределение мест по очкам")
plt.show()
```
