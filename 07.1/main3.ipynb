```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("eplleaguetables.csv")
```

```python
plt.figure(figsize=(15, 10))
ax = plt.subplot(111)
df[df["Position"] == 1]["Points"].plot(kind="bar", stacked=True, color="#C9F76F", ax=ax, label="1-3 место")
df[df["Position"] == 3]["Points"].plot(kind="bar", stacked=True, color="#5CCCCC", ax=ax, label="_place")
df[df["Position"] == 19]["Points"].plot(kind="bar", stacked=True, color="#FF7373", ax=ax, label="19-20 место")
df[df["Position"] == 20]["Points"].plot(kind="bar", stacked=True, color="#FFFFFF", ax=ax, label="_place",
                                        alpha=0.7)
ax.set_xticklabels(df["Season"].drop_duplicates(), rotation=0)
ax.set_xlabel("Сезон, гг")
ax.set_ylabel("Очки")
ax.set_yticks(list(range(df["Points"].min(), df["Points"].max(), 3)) + [df["Points"].max()])
ax.set_axisbelow(True)
ax.axhline(df[df["Position"] == 19]["Points"].max() + 1, color="black", linewidth=2, linestyle="--",
           label='"Безопасная" зона')
ax.legend()
ax.grid(axis="y", linestyle="--")
ax.set_title("Распределение мест по очкам")
plt.show()
```

По графику видно, что негативному красноватому цвету соответствуют очки команд 19 и 20 мест (и меньшие очки), а "безопасная" зона начинается с очков синеватого цвета, при наборе которых команда не вылетает в низший дивизион

### Ответ на вопрос задачи

Пунктирной линией отмечена линия наибольшего количества очков у 19 места + 1. Начиная с этого количества очков (40), команда не вылетает в низший дивизион во всех сезонах. Так как мы смотрим на все сезоны, то надо брать максимум, а не среднее значение, потому что иначе в каких=то сезонах все равно произошел бы вылет

**Ответ**: нужно набрать не менее 40 очков, чтобы гарантированно (т.е. во всех сезонах) не вылететь
